```python
#since car length and curbweight are highly correlated hence we take curbweight

df=df.drop(columns=['carlength'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Extract the features to be normalized into a separate DataFrame
feature_normalize = df[['carwidth','curbweight','enginesize','horsepower']]

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the features
scaler.fit(feature_normalize)

# Transform the features using the scaler
normalized_features = scaler.transform(feature_normalize)

# Replace the original features in the DataFrame with the normalized ones
df[['carwidth','curbweight','enginesize','horsepower']] = normalized_features
```

```python
df.head()
```

```python
data = df.values
x = data[:,0:4]
y = data[:,4]
```

```python
# train test split

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
ypred = lr.predict(xtest)
```

```python
lr.predict([[0.174483,0.074449,-0.844782,-0.014566]])
```

```python
from sklearn.metrics import r2_score
ac = r2_score(ytest,ypred)
ac
```
