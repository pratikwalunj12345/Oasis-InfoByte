```python
sns.pairplot(unemp)
```

```python
plt.bar(unemp['Region'], unemp['Estimated Unemployment Rate (%)'])
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment according to States')
plt.show()
```

```python
sns.histplot(x='Estimated Unemployment Rate (%)', data=unemp,kde = True,hue='Area')
plt.title('Unemployment according to Area')
plt.xlabel('Unemployment Rate')
plt.show()
```

```python
sns.lineplot(y='Estimated Unemployment Rate (%)', x='Date', data=unemp)
plt.title('Unemployment according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
plt.bar(unemp['Region'], unemp['Estimated Labour Participation Rate (%)'])
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Labour Participation Rate')
plt.title('Labour Participation according to States')
plt.show()
```

```python
sns.histplot(x='Estimated Labour Participation Rate (%)', data=unemp, kde=True, hue='Area')
plt.title('Labour Participation according to Area')
plt.xlabel('Labour Participation Rate')
plt.show()
```

```python
sns.lineplot(y='Estimated Labour Participation Rate (%)', x='Date', data=unemp)
plt.title('Labour Participation according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Labour Participation Rate')
plt.show()
```

```python
plt.bar(unemp['Region'], unemp['Estimated Employed'])
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Employment Rate')
plt.title('Employment according to States')
plt.show()
```

```python
sns.lineplot(y='Estimated Employed', x='Date', data=unemp)
plt.title('Employment according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Employment Rate')
plt.show()
```
